```python
import pandas as pd

def create_restrictions_column(df):
    # Create an empty list to store restriction column values
    restrictions = []

    # Iterate through each row
    for index, row in df.iterrows():
        # Initialize an empty list for the current row's restrictions
        row_restrictions = []
        
        # Iterate through each category column
        for col in df.columns:
            # Check if the column has a non-empty value
            if not pd.isnull(row[col]):
                # If a value exists in the column, add the column name to row_restrictions
                row_restrictions.append(col)
        
        # Join the row_restrictions list into a string, separated by commas, and append it to the restrictions list
        restrictions.append(', '.join(row_restrictions))
    
    # Add the restrictions list as a new column named "Restrictions" to the DataFrame
    df['Restrictions'] = restrictions
    
    return df

# Call the function to create the "Restrictions" column
df3 = create_restrictions_column(df3)

# Display the updated DataFrame
print(df3[['Name', 'Description', 'Restrictions']])
```

```python
# Define a function to download images for food names
def download_food_images(df, output_path):
    for index, row in df.iterrows():
        # Extract the food name from the DataFrame
        food_name = row["Name"]

        # Search for images related to the food name
        images = search_images_ddg(food_name)

        if len(images) > 0:
            # Create a subdirectory for each food name within the "Images" directory
            food_dir = output_path / food_name
            food_dir.mkdir(exist_ok=True)

            # Download and save each image
            for i, img_url in enumerate(images):
                img = PILImage.create(img_url)
                img.save(food_dir / f"{i}.jpg")

            print(f"Downloaded {len(images)} images for '{food_name}' to '{food_dir}'")
        else:
            print(f"No images found for '{food_name}'")

# Example usage:
# Assuming df is your DataFrame containing the "Food_name" column
# and you want to save the images in the "Images" directory
download_food_images(df, Path("Images"))
```

```python
import pandas as pd

# Assuming df is your DataFrame and 'Name' is your column
df = pd.read_csv('Dataset(final).csv')

# Characters that are not allowed in Windows filenames
invalid_chars = ['<', '>', ':', '"', '/', '\\', '|', '?', '*']

def clean_name(name):
    # Replace spaces and single quotes with underscores
    name = name.replace(' ', '_').replace("'", "_")
    
    # Remove any characters that are not allowed in Windows filenames
    for char in invalid_chars:
        name = name.replace(char, '')
    
    return name

df['Name'] = df['Name'].apply(clean_name)

print(df)
```

```python
def remove_duplicates(label_str):
    # Split the string into a list of labels
    labels = label_str.split(',')
    # Remove leading/trailing whitespace
    labels = [label.strip() for label in labels]
    # Remove duplicates by converting the list to a set, then convert it back to a list
    labels = list(set(labels))
    # Join the labels back into a string
    label_str = ', '.join(labels)
    return label_str

# Apply the function to the 'label' column
df['label'] = df['label'].apply(remove_duplicates)
```
